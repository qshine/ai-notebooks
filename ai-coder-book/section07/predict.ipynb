```python
import os
import cv2
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import to_categorical
from keras.models import load_model
```

```python
data_path = '/Users/xxx/Downloads/data/archive/'
num_classes = 43
```

```python
def get_test_data(csv_path, data_path):
    '''
    从csv中读取分类和图片路径
    '''
    test = pd.read_csv(csv_path, sep=',')
    X_test = []
    y_test = []

    i = 0
    for file_name, class_id in zip(list(test['Path']), list(test['ClassId'])):
        img_path = os.path.join(data_path, file_name)
        img = cv2.resize(cv2.imread(img_path), (32, 32))

        X_test.append(img)
        y_test.append(class_id)
        
        i = i + 1
        print('loaded image{}'.format(i))

    X_test = np.array(X_test)
    y_test = np.array(y_test)

    return X_test, y_test
```

```python
# 准备测试数据
test_x, test_y = get_test_data(data_path+'Test.csv', data_path)
test_labels = to_categorical(test_y, num_classes)
```

```python
# 重新加载模型
model = load_model('gtsrb_model.h5')

# 对测试数据进行预测
test_loss, test_acc = model.evaluate(test_x, test_labels, verbose=0)

# 输出准确率
print('Test accuracy:', test_acc)
```

```python
# 观察一部分预测数据
for i in range(10):
    x = test_x[i]
    y = test_y[i]
    pred_y = np.argmax(model.predict(np.array([x])))
    print('real => predict: {}={}'.format(y, pred_y))
```
